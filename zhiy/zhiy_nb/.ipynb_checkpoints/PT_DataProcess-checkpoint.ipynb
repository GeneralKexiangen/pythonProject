```python
datas = []
datas.append(row1)
datas.append('\n')
datas.append(row2)
datas.append(row3)
i =0 
for index,row in df.iterrows():
    i += 1
    datas.append(value1.format(row['IFIX变量名']+'_STATUS',row['PLC_设备名称']+'设备状态','BJZS.YPA.DB57，B'+str(row['起始地址'])))
    if i >2:
        break
datas.append('\n')
datas.append(row4)
datas.append(row5)
n = 0
for index,row in df.iterrows():
    n += 1
    for j in range(1,11):
        v1 = row['IFIX变量名']+tags[j-1]
        v2 = row['PLC_设备名称']+tags_name[j-1]
        if j <8:
            v3 = 'BJZS.YPA.DB57，B{0}.{1}'.format(str(int(row['起始地址'])+1),str(j-1))
        elif j ==8:
            v3 = 'BJZS.YPA.DB57，B{0}.{1}'.format(str(int(row['起始地址'])+3),'0')
        elif j ==9:
            v3 = 'BJZS.YPA.DB57，W{0}'.format(str(row['起始地址']))
        elif j ==10:
            v3 = 'BJZS.YPA.DB57，W{0}'.format(str(int(row['起始地址'])+2))
        datas.append(value2.format(v1,v2,v3))
    if n>2:
        break
datas.append('\n')
datas.append(row6)
datas
```

```python
import csv
import codecs
def data_write_csv(file_name, datas):#file_name为写入CSV文件的路径，datas为要写入数据列表
    file_csv = codecs.open(file_name,'w+','utf-8')#追加
    writer = csv.writer(file_csv, delimiter=' ', quotechar=' ', quoting=csv.QUOTE_MINIMAL)
    for data in datas:
        writer.writerow(data)
    print("保存文件成功，处理结束")
```

```python
data_write_csv('BJZS_EM.csv',datas)
```
