### Linear regression model selection 

```python
# minimize test loss

from sklearn.model_selection import train_test_split

target_y = n_news['popularity']
features_x = n_news['kw_min_avg']

X_train,X_test,y_train,y_test = train_test_split(features_x, target_y, test_size = 0.2, random_state = 1)
X_train,X_validation, y_train,y_validation = train_test_split(X_train, y_train, test_size = 0.2, random_state = 1)
```

```python
# 有多少训练集
X_train.shape[0]
```

```python
y_validation.shape[0]
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

max_deg = 8

train_error = list()
validation_error = list()

for i in range(1, max_deg):
    poly_transformer = PolynomialFeatures(i)
    poly_x = poly_transformer.fit_transform(X_train.values.reshape(-1, 1))
    poly_x_vali = poly_transformer.fit_transform(X_validation.values.reshape(-1, 1))
    lin_reg = LinearRegression()
    
    #estimate Coefficient
    lin_reg.fit(poly_x, y_train)
    
    y_pred = lin_reg.predict(poly_x)
    y_pred_vali = lin_reg.predict(poly_x_vali)
    
    mse_train = mean_squared_error(y_train, y_pred)
    train_error.append(mse_train)

    mse_valid = mean_squared_error(y_validation, y_pred_vali)
    validation_error.append(mse_valid)
```

```python
fig_1 = plt.figure()
plt.plot(np.arange(1, max_deg, 1), train_error, label = 'Train Error')
plt.plot(np.arange(1, max_deg, 1), validation_error, label = 'validation Error')

plt.title('Training Error vs Polynomial Degree')
plt.xlabel('Polynomial Degree')
plt.ylabel('Training Error')
plt.legend()
```

```python
from sklearn.model_selection import cross_val_score, LeaveOneOut, KFold
kf = KFold(10)

neighbours = np.arange(1, 101)

cv_rmse = []
for k in neighbours:
    knn = KNeighborsRegressor(n_neighbors = k)
    scores = cross_val_score(knn, news[['kw_min_avg']], news['popularity'], cv = kf, scoring = 'neg_mean_squared_error')
    rmse = np.sqrt(-1*np.mean(scores))
    cv_rmse.append(rmse)
    knn.fit(news[['kw_min_avg']], news['popularity'])
    
    
fig, ax = plt.subplots()
ax.plot(neighbours, cv_rmse, color = '#EF7F0E', label = 'CV error')
ax.set_xlabel("Neighbours numbers")
ax.set_ylabel("RMSE")
plt.legend()
plt.show()

print(1 + np.argmin(cv_rmse))
    
```
