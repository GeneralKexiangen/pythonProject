```python
df['销售月份']= df['销售日期'].apply(lambda x: str(x)[0:7])
df.head()
```

```python
mons = set(df['销售月份'])
mons
```

```python
cols =['销售月份','销售区域','销售额']
citys= ['上海', '北京', '南京', '安徽', '广东', '江苏', '浙江', '福建']
mons = ['2020-01','2020-02','2020-03','2020-04','2020-05','2020-06','2020-07','2020-08','2020-09','2020-10','2020-11','2020-12']
def get_df(brand):
    print(brand)
    brand_df = df[df['品牌']==brand][cols]
    brand_df = brand_df.groupby(['销售月份','销售区域']).agg('sum')
    brand_df = brand_df.reset_index()
    return brand_df
```

```python
brands
```

```python
get_df('九月风')
```

```python
def get_bdfs(brand):
    bdf =get_df(brand)
    adds = []
    for mon in mons:
        df_mon = bdf[bdf['销售月份']==mon]
        cts = set(df_mon['销售区域'])
        for ct in citys:
            if ct not in cts:
                adds.append([mon,ct,10000])
    add_df = pd.DataFrame(adds,columns=['销售月份','销售区域','销售额'])
    bdf = bdf.append(add_df, ignore_index=True)
    bdf = bdf.sort_values(by=['销售月份','销售区域'])
    return bdf
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

def fig_lines(bdf,bd):
    xlabels = []
    for x in bdf['销售月份']:
        if x not in xlabels:
            xlabels.append(x)
    print(xlabels)
    shanghai = list(bdf[bdf['销售区域']=='上海']['销售额'])
    beijing = list(bdf[bdf['销售区域']=='北京']['销售额'])
    nanjing = list(bdf[bdf['销售区域']=='南京']['销售额'])
    anhui = list(bdf[bdf['销售区域']=='安徽']['销售额'])
    guangdong = list(bdf[bdf['销售区域']=='广东']['销售额'])
    jiangsu = list(bdf[bdf['销售区域']=='江苏']['销售额'])
    zhejiang = list(bdf[bdf['销售区域']=='浙江']['销售额'])
    fujian = list(bdf[bdf['销售区域']=='福建']['销售额'])
    print(fujian)
    
#     params = {'figure.figsize': '8, 4'}
#     plt.rcParams.update(params)
    plt.figure(figsize=(30,5))
    matplotlib.rc("font", family='Heiti TC')
    barWidth = 0.1
    r1 = np.arange(12)
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]
    r5 = [x + barWidth for x in r4]
    r6 = [x + barWidth for x in r5]
    r7 = [x + barWidth for x in r6]
    r8 = [x + barWidth for x in r7]
    # 创建柱子 b---blue c---cyan g---green k----black m---magenta r---red w---white y----yellow
    plt.bar(r1, shanghai, color='b', width=barWidth, edgecolor='white', label='上海')
    plt.bar(r2, beijing, color='c', width=barWidth, edgecolor='white', label='北京')
    plt.bar(r3, nanjing, color='g', width=barWidth, edgecolor='white', label='南京')
    plt.bar(r4, anhui, color='k', width=barWidth, edgecolor='white', label='安徽')
    plt.bar(r5, guangdong, color='m', width=barWidth, edgecolor='white', label='广东')
    plt.bar(r6, jiangsu, color='r', width=barWidth, edgecolor='white', label='江苏')
    plt.bar(r7, zhejiang, color='#FF0088', width=barWidth, edgecolor='white', label='浙江')
    plt.bar(r8, fujian, color='y', width=barWidth, edgecolor='white', label='福建')

    # 添加x轴名称
    plt.xticks([r + barWidth for r in range(12)], xlabels)
    plt.xlabel('月份')
    plt.ylabel('销售额')
    plt.title('{0} 各地区销售额分布图'.format(bd))
    # 创建图例
    plt.legend()
    # 展示图片
    plt.show()
```

```python
for bd in brands:
    fig_lines(get_bdfs(bd),bd)
```
